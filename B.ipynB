```python
import spacy
import pandas as pd
 
data = pd.read_csv("IMDB Dataset.csv")
nlp = spacy.load("en_core_web_sm")
```

```python
adj = []
for num in range(50000):
    raw_text = nlp(data['review'][num])
    for token in raw_text:
        if(token.pos_ == 'ADJ'):
            adj.append(token.text)
adj = set(adj)
adj = list(adj)
print(adj)
```

```python
pos = []
neg = []

for words in adj:
    numOfPos = 0
    numOfNeg = 0
    for num in range(100):
        raw_text = data['review'][num]
        arr = raw_text.split(" ")
        for number in range(len(arr)):
            if(words == arr[number]):
                if(data['sentiment'][num] == 'positive'):
                    numOfPos += 1
                else:
                    numOfNeg += 1
    pos.append(numOfPos)
    neg.append(numOfNeg)
```

```python
dataa = []
for i in range(len(adj)):
    data2 = []
    data2.append(adj[i])
    data2.append(pos[i])
    data2.append(neg[i])
    dataa.append(data2)

df = pd.DataFrame(dataa, columns=['word','positive','negative'])
print(df)
```
